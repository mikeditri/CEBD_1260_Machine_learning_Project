```python
def dtype_conver(Dataframe):
    for col in Dataframe:
        if Dataframe[col].dtype == 'float':
            Dataframe[col] = Dataframe[col].astype(np.float16)
        if Dataframe[col].dtype == 'int64':
            Dataframe[col] = Dataframe[col].astype(np.int16)
```

```python
def mem_use(Dataframe):
    mem_use = Dataframe.memory_usage().sum() / 1024 ** 3  # this data set takes >1.7G RAM, we should optimize it after
    print('Memory usage of dataframe is {:.2f} GB'.format(mem_use))
```

```python
#Feature Engineering
train_df['Original_Quote_Date'] = pd.to_datetime(train_df['Original_Quote_Date'])
train_df['Original_Quote_Date'].dtype
train_df['year'] = train_df['Original_Quote_Date'].dt.year
train_df['month'] = train_df['Original_Quote_Date'].dt.month
train_df['day'] = train_df['Original_Quote_Date'].dt.day
```

```python
# define target
y = train_df["QuoteConversion_Flag"].values

# drop target column from data
X = train_df.drop(["QuoteConversion_Flag"], axis = 1)

#QuoteNumber setting as index
X = X.set_index("QuoteNumber")

```

```python
#select all numerical columns, then check if each column is indeed numerical
numerical_cols = [cols for cols in X.columns if train_df[cols].dtype in['int8','int16','int32','int64','float16',
                                                                               'float32','float64']]
#get min value of each numerical comlumn
minimum = X[numerical_cols].min().values

#get max value of each numerical comlumn
maximum = X[numerical_cols].max().values

#get number of nuniques per column
nuniques = X[numerical_cols].nunique().values

#display all rows
pd.set_option('display.max_rows', None)

#build dataframe of column min and max values using a directory
num_df = pd.DataFrame({'columns' :numerical_cols, 'min' :minimum, 'max' :maximum, 'nuniques' :nuniques})
```

```python
#check which fields to convert to onehot encoding then put all those that need encoding into a list
t = []
for col in X.columns:
    if X[col].name not in ['Original_Quote_Date', 'SalesField8']:
        t.append(X[col].name)
# convert all to onehot encoding and sparse matrix except 'Original_Quote_Date' & 'SalesField8'
```

```python
temp_df = pd.get_dummies(data=X, columns=t)
```
